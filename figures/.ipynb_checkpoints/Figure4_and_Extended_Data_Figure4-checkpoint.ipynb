```python
%pylab inline
import os
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
def find_variants(pval_file, start, end):
    x = []
    header = True
    for line in open(pval_file):
        if header:
            header = False
            continue
        elems = line.strip().split()
        pos = int(elems[1])
        if pos >= start and pos < end:
            x.append(pos)
    return x
           
def get_scores(pos_list, start, weights):
    y = []
    for pos in pos_list:
        pos_offset = pos-start
        y.append(weights[pos_offset])
    return y



def draw_figure(weight_file, position_file, pval_file, target_chrom, target_pos, target_score, fig_idx, shorten):
    score_map = {}
    with open(weight_file) as f, open(position_file) as f_pos:
        while True:
            weight_info = f.readline().strip()
            position_info = f_pos.readline().strip()

            if (weight_info != "") and (position_info != ""):
                chromID, motif_start, motif_end, seq_start, seq_end, strand = position_info.split(",")
                seq_start, seq_end = int(seq_start), int(seq_end)
                if (chromID == target_chrom) and (seq_start <= target_pos) and (seq_end > target_pos):
                    seq_start_plot = seq_start
                    seq_end_plot = seq_end
                    weights_raw = [abs(float(wt)) for wt in weight_info.split(";")]
                    weights_sum = sum(weights_raw)
                    weights = [w/weights_sum for w in weights_raw]
                    for idx in range(1000):
                        if (chromID, seq_start+idx) not in score_map:
                            score_map[(chromID, seq_start+idx)] = 0

                        if strand == "+":
                            if abs(weights[idx]) > abs(score_map[(chromID, seq_start+idx)]):
                                score_map[(chromID, seq_start+idx)] = weights[idx]
                        elif strand == "-":
                            if abs(weights[999-idx]) > abs(score_map[(chromID, seq_start+idx)]):
                                score_map[(chromID, seq_start+idx)] = weights[999-idx]
                        else:
                            exit("wrong strand symbol: %s" %(strand))
            elif (weight_info == "") and (position_info == ""):
                break
            else:
                exit("Files do not match:\n%s\n%s" %(weight_file, position_file))
               
    print("%s:%d-%d" %(target_chrom, seq_start_plot, seq_end_plot))
    weights_plot = [score_map[(target_chrom, seq_start_plot+idx)] for idx in range(1000)]
    variants_pos = find_variants(pval_file, seq_start_plot, seq_end_plot)
    variants_scores = get_scores(variants_pos, seq_start_plot, weights_plot)
                        
    fig = plt.figure()
    fig.set_size_inches((16,9))
    ax = fig.add_subplot(111)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.bar([pos+seq_start_plot for pos in range(1000)], weights_plot, width=1, color="darkgray", zorder=0)
    
    ax.bar(variants_pos, variants_scores, width=5, color="goldenrod", zorder=10)
    ax.scatter(variants_pos, [scr + max(weights_plot)/30.0 for scr in variants_scores], color="goldenrod", marker=11, zorder=11)
    
    print(get_scores([target_pos], seq_start_plot, weights_plot))
    ax.bar([target_pos], get_scores([target_pos], seq_start_plot, weights_plot), width=5, color="crimson", zorder=20)
    ax.scatter([target_pos], [scr + max(weights_plot)/30.0 for scr in get_scores([target_pos], seq_start_plot, weights_plot)],
               color="crimson", marker=11, zorder=21)
    ax.set_xlabel("Position", size=12, fontname="Arial")
    ax.set_ylabel("Importance score", size=12, fontname="Arial")
    ax.ticklabel_format(useOffset=False, style='plain', axis="x")
    ax.ticklabel_format(style='scientific', axis="y", scilimits=(0, 0))
    ax.set_xlim(seq_start_plot, seq_end_plot)
    plt.tight_layout()
    #fig.savefig("./importance_scores_%d.pdf" %(fig_idx))
            
    x = []
    y = []
    header = True
    for line in open(pval_file):
        if header:
            header = False
            continue
        elems = line.strip().split()
        chromID_pval = elems[0]
        pos_pval = int(elems[1])
        pval = float(elems[6])            
        if pval == 0:
            pval_log = 0
        else:
            pval_log = -math.log10(pval)
        if target_pos == pos_pval and chromID_pval == target_chrom:
            target_pval = pval_log
        x.append(pos_pval)
        y.append(pval_log)
        
    fig = plt.figure()
    fig.set_size_inches((16, 9))
    ax = fig.add_subplot(111)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.scatter(x, y, marker=".", color="darkgray")
    ax.scatter([target_pos], [target_pval], marker="*", color="crimson", s=200)
    ax.set_xlabel("Position", size=12, fontname="Arial")
    ax.set_ylabel("GWAS log10 p-value", size=12, fontname="Arial")
    if not shorten:
        ax.set_xlim(min(x)-1000, max(x)+1000)
    else:
        ax.set_xlim((min(x)+max(x))//2-126000, (min(x)+max(x))//2+126000)
    ax.ticklabel_format(style='plain', axis="x")
    #fig.savefig("./manhattan_%d.pdf" %(fig_idx))
```

```python
def draw_main(TF_name, chromID, pos, ab_score, idx, pval_file, shorten=False):
    data_path = "/storage/pandaman/project/Alzheimers_ResNet/storage/experiments/"
    position_file = os.path.join(data_path, "seqs_one_hot_extended/%s/visualization/auxiliary_info.txt" %(TF_name))
    weight_file = os.path.join(data_path, "results_extended_coordconv_sliding/%s/annations/scores.txt" %(TF_name))
    draw_figure(weight_file, position_file, pval_file, chromID, pos, ab_score, idx, shorten)
```
